```python
import os, math
import numpy as np
import cv2
import matplotlib.pyplot as plt
from collections import defaultdict

# ============================================================
#  Basic helpers
# ============================================================

# side indices: 0=top, 1=right, 2=bottom, 3=left
OPP = {0: 2, 1: 3, 2: 0, 3: 1}

def lerp(a, b, t):
    """Linear interpolation"""
    return a + (b - a) * t

def ease_in_out(t):
    """Smooth ease-in-out curve"""
    return t * t * (3 - 2 * t)

def rot90n(img, n):
    n %= 4
    if n == 0: return img
    if n == 1: return cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    if n == 2: return cv2.rotate(img, cv2.ROTATE_180)
    return cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

# ---------- RGB (.rgb) reader with auto layout detection ----------

def read_rgb_raw_auto(path):
    size   = os.path.getsize(path)
    pixels = size // 3
    side   = int(round(pixels ** 0.5))

    with open(path, "rb") as f:
        raw = np.frombuffer(f.read(), dtype=np.uint8)

    # candidate 1: interleaved H×W×3
    inter = raw.reshape((side, side, 3))
    inter_bgr = cv2.cvtColor(inter, cv2.COLOR_RGB2BGR)

    # candidate 2: planar 3×H×W
    planar = raw.reshape((3, side, side)).transpose(1, 2, 0)
    planar_bgr = cv2.cvtColor(planar, cv2.COLOR_RGB2BGR)

    def colorfulness(bgr):
        b, g, r = cv2.split(bgr.astype(np.float32))
        return float(np.mean(np.std(np.stack([b, g, r], axis=-1), axis=-1)))

    return planar_bgr if colorfulness(planar_bgr) > colorfulness(inter_bgr) else inter_bgr

def read_image_auto(path):
    ext = os.path.splitext(path.lower())[1]
    if ext == ".rgb":
        return read_rgb_raw_auto(path)
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    if img is None:
        raise FileNotFoundError(path)
    return img

# ============================================================
#  Segmentation + deskew with masks (aliasing-safe)
# ============================================================

def segment_pieces(bgr, thresh=None, min_area=300):
    """
    Segment rectangular pieces against dark background.

    If thresh is None, choose it adaptively from darkest 5% of pixels.
    """
    gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)

    if thresh is None:
        flat = gray.reshape(-1)
        p5   = np.percentile(flat, 5)
        thr  = p5 + 3
        thr  = max(1, min(50, thr))
    else:
        thr = thresh

    mask = (gray > thr).astype(np.uint8)

    kernel = np.ones((3,3), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)

    num, labels, stats, _ = cv2.connectedComponentsWithStats(mask, 8)

    pieces = []
    for i in range(1, num):
        x, y, w, h, area = stats[i]
        if area < min_area:
            continue
        crop  = bgr[y:y+h, x:x+w]
        cmask = (labels[y:y+h, x:x+w] == i).astype(np.uint8)
        pieces.append({"crop": crop, "mask": cmask, "bbox": (x, y, w, h)})
    return pieces

def deskew_piece_masked(piece):
    """
    piece: {"crop": bgr, "mask": uint8}
    Returns upright_img, upright_mask (tightly cropped).
    """
    crop = piece["crop"]
    mask = piece["mask"]

    ys, xs = np.where(mask > 0)
    pts = np.vstack([xs, ys]).T.astype(np.float32)
    rect = cv2.minAreaRect(pts)
    angle, (w, h) = rect[2], rect[1]

    # force more "upright" orientation
    if w < h:
        angle += 90.0

    center = (crop.shape[1] // 2, crop.shape[0] // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)

    rot  = cv2.warpAffine(crop, M, (crop.shape[1], crop.shape[0]),
                          flags=cv2.INTER_LINEAR,
                          borderValue=(0,0,0))
    rmask = cv2.warpAffine(mask, M, (mask.shape[1], mask.shape[0]),
                           flags=cv2.INTER_NEAREST,
                           borderValue=0)

    ys, xs = np.where(rmask > 0)
    y0, y1 = ys.min(), ys.max()+1
    x0, x1 = xs.min(), xs.max()+1

    #return rot[y0:y1, x0:x1], rmask[y0:y1, x0:x1], angle  # Enable this when deskew is active
    return crop, mask, 0.0  # Deskew disabled: return 0 since no rotation was applied

# ============================================================
#  Edge signatures (mask-aware)
# ============================================================

def edge_signature_masked(bgr, mask, side, strip=3, N=128):
    """
    Build a 1D color+gradient signature along a side, using only mask==1 pixels
    to avoid aliasing/background.
    side: 0=top,1=right,2=bottom,3=left
    """
    h, w, _ = bgr.shape
    strip2 = min(strip*2, h if side in (0,2) else w)

    line_cols = []

    if side == 0:       # top
        for x in range(w):
            band_m = mask[0:strip2, x]
            band_c = bgr[0:strip2, x]
            if np.any(band_m):
                pix = band_c[band_m.astype(bool)]
            else:
                pix = band_c.reshape(-1,3)
            line_cols.append(pix.mean(axis=0))
    elif side == 2:     # bottom
        for x in range(w):
            band_m = mask[h-strip2:h, x]
            band_c = bgr[h-strip2:h, x]
            if np.any(band_m):
                pix = band_c[band_m.astype(bool)]
            else:
                pix = band_c.reshape(-1,3)
            line_cols.append(pix.mean(axis=0))
    elif side == 1:     # right
        for y in range(h):
            band_m = mask[y, w-strip2:w]
            band_c = bgr[y, w-strip2:w]
            if np.any(band_m):
                pix = band_c[band_m.astype(bool)]
            else:
                pix = band_c.reshape(-1,3)
            line_cols.append(pix.mean(axis=0))
    else:               # left
        for y in range(h):
            band_m = mask[y, 0:strip2]
            band_c = bgr[y, 0:strip2]
            if np.any(band_m):
                pix = band_c[band_m.astype(bool)]
            else:
                pix = band_c.reshape(-1,3)
            line_cols.append(pix.mean(axis=0))

    line = np.asarray(line_cols, dtype=np.uint8)   # (L,3)

    # convert to Lab for better perceptual distance
    lab = cv2.cvtColor(line.reshape(-1,1,3), cv2.COLOR_BGR2LAB).reshape(-1,3).astype(np.float32)

    src = np.linspace(0, len(lab)-1, len(lab), dtype=np.float32)
    dst = np.linspace(0, len(lab)-1, N,          dtype=np.float32)
    res = np.vstack([np.interp(dst, src, lab[:,c]) for c in range(3)]).T  # (N,3)

    grad = np.gradient(res, axis=0)
    sig  = np.hstack([res, grad])   # (N,6)
    return sig

# ============================================================
#  Load puzzle pieces + rotations + signatures (MASK VERSION)
# ============================================================

def load_puzzle_pieces_masked(path, Nsig=128):
    print("Reading input...")
    bgr = read_image_auto(path)
    original_canvas = bgr.copy()  # Preserve for animation

    print("Segmenting pieces...")
    comps = segment_pieces(bgr)
    print(f"Found {len(comps)} pieces.")

    print("Deskewing pieces (with masks)...")
    uprights, masks = [], []
    original_bboxes = []
    original_crops = []
    deskew_angles = []
    for p in comps:
        img_u, mask_u, angle = deskew_piece_masked(p)
        uprights.append(img_u)
        masks.append(mask_u)
        original_bboxes.append(p["bbox"])
        original_crops.append(p["crop"])
        deskew_angles.append(angle)

    print("Building rotations + mask-aware signatures...")
    rotations        = []
    rotations_masks  = []
    sigs             = {}

    for i, (img_u, mask_u) in enumerate(zip(uprights, masks)):
        rots = []
        rmsk = []
        for r in range(4):
            rots.append(rot90n(img_u, r))
            rmsk.append(rot90n(mask_u, r))
        rotations.append(rots)
        rotations_masks.append(rmsk)

        for r in range(4):
            img_r = rots[r]
            msk_r = rmsk[r]
            for s in range(4):
                sigs[(i,r,s)] = edge_signature_masked(img_r, msk_r, s, strip=3, N=Nsig)
        
    print("Done preprocessing (aliasing-safe).")
    
    # Bundle metadata for animation
    piece_metadata = {
        "original_canvas": original_canvas,
        "original_bboxes": original_bboxes,
        "original_crops": original_crops,
        "deskew_angles": deskew_angles,
    }
    
    return uprights, masks, rotations, rotations_masks, sigs, piece_metadata

# ============================================================
#  Border detection (rotation-invariant)
# ============================================================

def infer_grid_dims(N):
    best = None
    for r in range(1, int(math.sqrt(N))+1):
        if N % r == 0:
            c = N // r
            if best is None or abs(r-c) < abs(best[0]-best[1]):
                best = (r,c)
    return best

def edge_distance(sigA, sigB):
    return float(np.linalg.norm(sigA - sigB[::-1]))

def compute_outerness(rotations, sigs):
    keys = list(sigs.keys())  # (i,r,s)
    outer = {}
    vals  = []
    for (i,ri,si) in keys:
        A = sigs[(i,ri,si)]
        best = np.inf
        for (j,rj,sj) in keys:
            if i == j: 
                continue
            if sj != OPP[si]:
                continue
            d = float(np.linalg.norm(A - sigs[(j,rj,sj)][::-1]))
            if d < best:
                best = d
        outer[(i,ri,si)] = best
        vals.append(best)

    vals = np.array(vals, dtype=np.float32)
    med  = np.median(vals)
    mad  = np.median(np.abs(vals - med)) + 1e-8
    thr  = med + 2.5 * mad

    border_mask = {k: (outer[k] > thr) for k in outer}
    return outer, border_mask, thr

def classify_pieces(rotations, border_mask):
    N = len(rotations)
    border_sides = {}
    roles = {}
    for i in range(N):
        votes = {c:0 for c in range(4)}
        for r in range(4):
            for s in range(4):
                if border_mask[(i,r,s)]:
                    c = (s - r) % 4
                    votes[c] += 1
        canonical_is_border = {c:(votes[c]>0) for c in range(4)}

        for r in range(4):
            sides_r = set()
            for s in range(4):
                c = (s - r) % 4
                if canonical_is_border[c]:
                    sides_r.add(s)
            border_sides[(i,r)] = sides_r

        n_border = sum(1 for c in range(4) if canonical_is_border[c])
        if n_border >= 2:
            roles[i] = "corner"
        elif n_border == 1:
            roles[i] = "edge"
        else:
            roles[i] = "interior"
    return roles, border_sides

def outward_sides_for_cell(r, c, R, C):
    sides = set()
    if r == 0:   sides.add(0)
    if c == C-1: sides.add(1)
    if r == R-1: sides.add(2)
    if c == 0:   sides.add(3)
    return sides

def rotation_ok_for_cell(i, rot, r, c, R, C, border_sides):
    outward = outward_sides_for_cell(r, c, R, C)
    sides   = border_sides[(i, rot)]
    for s in outward:
        if s not in sides:
            return False
    return True

def inward_border_penalty(i, rot, r, c, R, C, border_sides, gamma=30.0):
    outward = outward_sides_for_cell(r, c, R, C)
    sides   = border_sides[(i, rot)]
    inward  = {0,1,2,3} - outward
    n_bad   = sum(1 for s in inward if s in sides)
    return gamma * n_bad

# ============================================================
#  Best pair orientation + MST
# ============================================================

def best_pair_between(i, j, sigs):
    best = (1e18, 0,0,0,0)
    for ri in range(4):
        for si in range(4):
            sj = OPP[si]
            A  = sigs[(i,ri,si)]
            for rj in range(4):
                d = edge_distance(A, sigs[(j,rj,sj)])
                if d < best[0]:
                    best = (d, ri, si, rj, sj)
    return best

def kruskal_mst(N, edges):
    parent = list(range(N))
    rank   = [0]*N

    def find(x):
        while parent[x] != x:
            parent[x] = parent[parent[x]]
            x = parent[x]
        return x

    def union(a,b):
        ra, rb = find(a), find(b)
        if ra == rb:
            return False
        if rank[ra] < rank[rb]:
            parent[ra] = rb
        elif rank[ra] > rank[rb]:
            parent[rb] = ra
        else:
            parent[rb] = ra
            rank[ra]  += 1
        return True

    mst = []
    for w,i,j in sorted(edges, key=lambda e:e[0]):
        if union(i,j):
            mst.append((w,i,j))
            if len(mst) == N-1:
                break
    return mst

# ============================================================
#  Seam cost + total cost + uniform rendering
# ============================================================

def seam_cost_between_cells(grid, rotations, sigs, r1, c1, r2, c2):
    (i1, rot1) = grid[r1][c1]
    (i2, rot2) = grid[r2][c2]

    if r1 == r2:
        if c2 == c1+1:
            s1, s2 = 1, 3
        else:
            s1, s2 = 3, 1
    elif c1 == c2:
        if r2 == r1+1:
            s1, s2 = 2, 0
        else:
            s1, s2 = 0, 2
    else:
        return 0.0

    sig1 = sigs[(i1, rot1, s1)]
    sig2 = sigs[(i2, rot2, s2)]
    return float(np.linalg.norm(sig1 - sig2[::-1]))

def total_grid_seam_cost(grid, rotations, sigs):
    R, C = len(grid), len(grid[0])
    total = 0.0
    for r in range(R):
        for c in range(C-1):
            total += seam_cost_between_cells(grid, rotations, sigs, r, c, r, c+1)
    for r in range(R-1):
        for c in range(C):
            total += seam_cost_between_cells(grid, rotations, sigs, r, c, r+1, c)
    return total

def render_grid_uniform(grid, rotations, out_path=None, show=True, title=""):
    R, C = len(grid), len(grid[0])

    hs, ws = [], []
    for r in range(R):
        for c in range(C):
            i, rot_idx = grid[r][c]
            h, w = rotations[i][rot_idx].shape[:2]
            hs.append(h); ws.append(w)
    H0 = int(np.median(hs))
    W0 = int(np.median(ws))

    canvas = np.zeros((R*H0, C*W0, 3), np.uint8)
    for rr in range(R):
        for cc in range(C):
            i, rot_idx = grid[rr][cc]
            img = rotations[i][rot_idx]
            patch = cv2.resize(img, (W0, H0), interpolation=cv2.INTER_CUBIC)
            y0 = rr * H0
            x0 = cc * W0
            canvas[y0:y0+H0, x0:x0+W0] = patch

    if out_path:
        os.makedirs(os.path.dirname(out_path), exist_ok=True)
        cv2.imwrite(out_path, canvas)
    if show:
        plt.figure(figsize=(6,6))
        plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.title(title)
        plt.show()
    return canvas

def render_solution(solution, pieces, out_path=None, show=True, title=""):
    """
    Render solution from solve() which returns [[piece_idx, rotation, [row, col]], ...]
    where [row, col] are PIXEL coordinates (not grid cell indices).
    
    Args:
        solution: [[piece_idx, rotation, [row, col]], ...]
        pieces: Original pieces list (uprights)
        out_path: Output file path
        show: Whether to display
        title: Plot title
    """
    if not solution:
        print("No solution to render")
        return None
    
    piece_images = []
    positions = []
    
    for item in solution:
        piece_idx, rotation, pos = item
        piece_img = pieces[piece_idx]
        
        # Apply rotation if non-zero
        if rotation != 0:
            piece_img = rot90n(piece_img, rotation)
        
        piece_images.append(piece_img)
        
        # Ensure position is a list of 2 ints
        if isinstance(pos, np.ndarray):
            pos = pos.tolist()
        positions.append([int(pos[0]), int(pos[1])])
    
    # Find canvas bounds based on pixel positions + piece sizes
    min_row = min(p[0] for p in positions)
    min_col = min(p[1] for p in positions)
    max_row = max(p[0] + piece_images[i].shape[0] for i, p in enumerate(positions))
    max_col = max(p[1] + piece_images[i].shape[1] for i, p in enumerate(positions))
    
    # Canvas size in pixels
    canvas_h = max_row - min_row
    canvas_w = max_col - min_col
    
    # Create canvas
    canvas = np.zeros((canvas_h, canvas_w, 3), np.uint8)
    
    # Place pieces at their pixel coordinates
    for piece_img, pos in zip(piece_images, positions):
        # Offset positions so min_row/min_col maps to 0,0
        y0 = pos[0] - min_row
        x0 = pos[1] - min_col
        
        h, w = piece_img.shape[:2]
        canvas[y0:y0+h, x0:x0+w] = piece_img
    
    if out_path:
        out_dir = os.path.dirname(out_path)
        if out_dir:
            os.makedirs(out_dir, exist_ok=True)
        cv2.imwrite(out_path, canvas)
    if show:
        plt.figure(figsize=(6, 6))
        plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.title(title)
        plt.show()
    
    return canvas

# ============================================================
#  Rectangular MST assembly (tolerant size penalty)
# ============================================================

def assemble_rectangular_multi_seed(
        rotations, sigs, roles, border_sides,
        best_pairs, mst_adj,
        lam_mst=80.0, gamma=30.0,
        lam_size_h=40.0, lam_size_w=40.0,
        tol_h=2, tol_w=2):
    """
    N pieces -> rectangular grid R×C (R*C=N, as square-ish as possible).

    Size penalty is used only when size differences exceed (tol_h, tol_w):
      - small differences (e.g. 100 vs 101) are ignored,
      - big differences (e.g. 80 vs 120) are penalized to enforce
        real rectangle consistency.
    """
    N = len(rotations)
    R, C = infer_grid_dims(N)
    print(f"Using grid {R} x {C} for N={N}")

    def make_local_cell_cost(row_target_h, col_target_w):
        def local_cell_cost(grid, i, rot, rr, cc):
            cost  = 0.0
            bonus = 0.0

            # ----- seam with top neighbor -----
            if rr > 0 and grid[rr-1][cc] is not None:
                j, rj = grid[rr-1][cc]
                sig1 = sigs[(j, rj, 2)]
                sig2 = sigs[(i, rot, 0)]
                d    = float(np.linalg.norm(sig1 - sig2[::-1]))
                cost += d

                if j in mst_adj[i]:
                    bonus += 10.0

                bp = best_pairs.get((j,i), None)
                if bp is not None:
                    ri_best, si_best, rj_best, sj_best, d_best = bp
                    if (rj,2,rot,0) == (ri_best, si_best, rj_best, sj_best):
                        bonus += lam_mst

            # ----- seam with left neighbor -----
            if cc > 0 and grid[rr][cc-1] is not None:
                j, rj = grid[rr][cc-1]
                sig1 = sigs[(j, rj, 1)]
                sig2 = sigs[(i, rot, 3)]
                d    = float(np.linalg.norm(sig1 - sig2[::-1]))
                cost += d

                if j in mst_adj[i]:
                    bonus += 10.0

                bp = best_pairs.get((j,i), None)
                if bp is not None:
                    ri_best, si_best, rj_best, sj_best, d_best = bp
                    if (rj,1,rot,3) == (ri_best, si_best, rj_best, sj_best):
                        bonus += lam_mst

            # ----- inward border penalty -----
            cost += inward_border_penalty(i, rot, rr, cc, R, C, border_sides,
                                          gamma=gamma)

            # ----- tolerant size penalties -----
            h, w = rotations[i][rot].shape[:2]

            if row_target_h[rr] is not None:
                dh = abs(h - row_target_h[rr])
                if dh > tol_h:
                    cost += lam_size_h * (dh - tol_h)

            if col_target_w[cc] is not None:
                dw = abs(w - col_target_w[cc])
                if dw > tol_w:
                    cost += lam_size_w * (dw - tol_w)

            return cost - bonus
        return local_cell_cost

    # ----- candidate seeds for top-left -----
    seeds = []
    for i in range(N):
        for r in range(4):
            if rotation_ok_for_cell(i, r, 0, 0, R, C, border_sides):
                seeds.append((i,r))
    if not seeds:
        seeds = [(i,r) for i in range(N) for r in range(4)]
    print(f"Trying {len(seeds)} seeds for top-left")

    best_grid = None
    best_cost = float("inf")
    best_seed = None

    # ----- try each seed -----
    for seed in seeds:
        (i0, r0) = seed
        grid = [[None]*C for _ in range(R)]
        used = set([i0])
        grid[0][0] = (i0, r0)

        row_target_h = [None]*R
        col_target_w = [None]*C
        h0, w0 = rotations[i0][r0].shape[:2]
        row_target_h[0] = h0
        col_target_w[0] = w0

        local_cell_cost = make_local_cell_cost(row_target_h, col_target_w)

        for rr in range(R):
            for cc in range(C):
                if rr == 0 and cc == 0:
                    continue

                best = (1e18, None)

                # 1st pass: respect border classification
                for i in range(N):
                    if i in used:
                        continue
                    for rot in range(4):
                        if not rotation_ok_for_cell(i, rot, rr, cc, R, C, border_sides):
                            continue
                        v = local_cell_cost(grid, i, rot, rr, cc)
                        if v < best[0]:
                            best = (v, (i,rot))

                # fallback: ignore border if nothing fits
                if best[1] is None:
                    for i in range(N):
                        if i in used:
                            continue
                        for rot in range(4):
                            v = local_cell_cost(grid, i, rot, rr, cc)
                            if v < best[0]:
                                best = (v, (i,rot))

                i_best, rot_best = best[1]
                grid[rr][cc] = (i_best, rot_best)
                used.add(i_best)

                h, w = rotations[i_best][rot_best].shape[:2]
                if row_target_h[rr] is None:
                    row_target_h[rr] = h
                if col_target_w[cc] is None:
                    col_target_w[cc] = w

        cost = total_grid_seam_cost(grid, rotations, sigs)
        if cost < best_cost:
            best_cost = cost
            best_grid = grid
            best_seed = seed

    print("Best seed:", best_seed, "with seam cost:", best_cost)
    return best_grid


# ============================================================
#  Animation
# ============================================================

def animate_puzzle_assembly(
    solution,
    uprights,
    piece_metadata,
    duration=3.0,
    fps=30,
    out_path="puzzle_animation.mp4"
):
    """
    Animate puzzle assembly from scattered pieces to solved state.
    Supports both translation-only and rotation puzzles.
    
    Parameters
    ----------
    solution : List[List]
        [[piece_idx, rotation, [row, col]], ...] from solve()
    uprights : List[np.ndarray]
        Upright piece images (used for final rotated rendering)
    piece_metadata : dict
        Contains original_canvas, original_bboxes, original_crops, deskew_angles
    duration : float
        Animation duration in seconds
    fps : int
        Frames per second
    out_path : str
        Output MP4 file path
    """
    
    original_canvas = piece_metadata["original_canvas"]
    original_bboxes = piece_metadata["original_bboxes"]
    original_crops = piece_metadata["original_crops"]
    deskew_angles = piece_metadata["deskew_angles"]
    
    n_frames = int(duration * fps)
    
    # Extract positions and compute grid bounds
    positions = np.array([item[2] for item in solution])
    min_row, max_row = positions[:, 0].min(), positions[:, 0].max()
    min_col, max_col = positions[:, 1].min(), positions[:, 1].max()
    R = max_row - min_row + 1
    C = max_col - min_col + 1
    
    # Compute uniform final cell size
    final_sizes = []
    for piece_idx, rotation, _ in solution:
        img = rot90n(uprights[piece_idx], rotation)
        final_sizes.append((img.shape[0], img.shape[1]))
    
    hs = [s[0] for s in final_sizes]
    ws = [s[1] for s in final_sizes]
    H0 = int(np.median(hs))
    W0 = int(np.median(ws))
    
    final_canvas_h = R * H0
    final_canvas_w = C * W0
    
    # Canvas size: max of original and final
    canvas_h = max(original_canvas.shape[0], final_canvas_h)
    canvas_w = max(original_canvas.shape[1], final_canvas_w)
    
    # Offset to center final grid
    offset_y = (canvas_h - final_canvas_h) // 2
    offset_x = (canvas_w - final_canvas_w) // 2
    
    # Build trajectories with 5 parameters: (cx, cy, theta, w, h)
    trajectories = []
    
    for piece_idx, rotation, pos in solution:
        row, col = pos[0], pos[1]
        
        # Original state (from segmentation)
        x0, y0, w_orig, h_orig = original_bboxes[piece_idx]
        
        # Start: original position, no rotation, original size
        cx_start = x0 + w_orig / 2
        cy_start = y0 + h_orig / 2
        theta_start = 0.0  # Crop shows original visual orientation
        w_start = float(w_orig)
        h_start = float(h_orig)
        
        # End: grid position, rotated, uniform size
        cx_end = offset_x + (col - min_col) * W0 + W0 / 2
        cy_end = offset_y + (row - min_row) * H0 + H0 / 2
        
        # Total rotation = grid rotation (CW) minus deskew angle (CCW in preprocessing)
        # Currently: deskew_angles[i] ≈ 0 (disabled), rotation = 0 (translation-only)
        theta_end = rotation * 90.0 - deskew_angles[piece_idx]
        
        w_end = float(W0)
        h_end = float(H0)
        
        trajectories.append({
            "piece_idx": piece_idx,
            "original_crop": original_crops[piece_idx],
            "start": (cx_start, cy_start, theta_start, w_start, h_start),
            "end": (cx_end, cy_end, theta_end, w_end, h_end),
        })
    
    def render_frame(t):
        """Render a single frame at normalized time t ∈ [0, 1]"""
        frame = np.zeros((canvas_h, canvas_w, 3), dtype=np.uint8)
        
        for traj in trajectories:
            # Smooth interpolation
            local_t = ease_in_out(t)
            
            # Interpolate all 5 parameters
            cx = lerp(traj["start"][0], traj["end"][0], local_t)
            cy = lerp(traj["start"][1], traj["end"][1], local_t)
            theta = lerp(traj["start"][2], traj["end"][2], local_t)
            w = lerp(traj["start"][3], traj["end"][3], local_t)
            h = lerp(traj["start"][4], traj["end"][4], local_t)
            
            crop = traj["original_crop"]
            
            # Resize to interpolated size
            target_w = max(1, int(round(w)))
            target_h = max(1, int(round(h)))
            resized = cv2.resize(crop, (target_w, target_h), 
                                 interpolation=cv2.INTER_LINEAR)
            
            # Rotate around center (negative theta for cv2 convention)
            M = cv2.getRotationMatrix2D((target_w/2, target_h/2), -theta, 1.0)
            
            # Compute rotated bounding box
            cos_a = abs(M[0, 0])
            sin_a = abs(M[0, 1])
            new_w = int(target_h * sin_a + target_w * cos_a)
            new_h = int(target_h * cos_a + target_w * sin_a)
            
            # Adjust matrix for expanded canvas
            M[0, 2] += (new_w - target_w) / 2
            M[1, 2] += (new_h - target_h) / 2
            
            rotated = cv2.warpAffine(resized, M, (new_w, new_h), 
                                     borderValue=(0, 0, 0))
            
            # Place on canvas centered at (cx, cy)
            x1 = int(cx - new_w / 2)
            y1 = int(cy - new_h / 2)
            x2, y2 = x1 + new_w, y1 + new_h
            
            # Clip to canvas bounds
            src_x1 = max(0, -x1)
            src_y1 = max(0, -y1)
            src_x2 = new_w - max(0, x2 - canvas_w)
            src_y2 = new_h - max(0, y2 - canvas_h)
            
            dst_x1 = max(0, x1)
            dst_y1 = max(0, y1)
            dst_x2 = min(canvas_w, x2)
            dst_y2 = min(canvas_h, y2)
            
            if dst_x2 > dst_x1 and dst_y2 > dst_y1:
                patch = rotated[src_y1:src_y2, src_x1:src_x2]
                # Composite non-black pixels
                mask = np.any(patch > 0, axis=2)
                frame[dst_y1:dst_y2, dst_x1:dst_x2][mask] = patch[mask]
        
        return frame
    
    # Initialize video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out_dir = os.path.dirname(out_path)
    if out_dir:
        os.makedirs(out_dir, exist_ok=True)
    writer = cv2.VideoWriter(out_path, fourcc, fps, (canvas_w, canvas_h))
    
    # Generate and write frames
    print(f"Generating {n_frames} frames for animation...")
    for frame_num in range(n_frames):
        t = frame_num / (n_frames - 1) if n_frames > 1 else 1.0
        frame = render_frame(t)
        writer.write(frame)
    
    writer.release()
    print(f"Animation saved to {out_path}")

# ============================================================
#  Top-level solver
# ============================================================

def solve_puzzle_mst_rectangular(path, out_dir="./outputs_mst_rect", show=True, animate=False, anim_duration=3.0):
    print(f"=== Solving: {path} ===")
    uprights, masks, rotations, rotations_masks, sigs, piece_metadata = \
        load_puzzle_pieces_masked(path)
    N = len(rotations)
    print(f"{os.path.basename(path)}: {N} pieces")


    solution = solve(uprights)

    for p in range(len(solution)):
        idx, rot, pos = solution[p]
        print(f"piece {idx} (rot={rot}): {pos}")
    
    """
    outer, border_mask, thr = compute_outerness(rotations, sigs)
    roles, border_sides     = classify_pieces(rotations, border_mask)
    print("Border threshold:", thr)

    # piece-level edges for MST + best oriented contact between each pair
    edges      = []
    best_pairs = {}
    for i in range(N):
        for j in range(i+1, N):
            d,ri,si,rj,sj = best_pair_between(i, j, sigs)
            edges.append((d, i, j))
            best_pairs[(i,j)] = (ri,si,rj,sj,d)
            best_pairs[(j,i)] = (rj,sj,ri,si,d)

    mst = kruskal_mst(N, edges)
    mst_adj = defaultdict(set)
    for w,i,j in mst:
        mst_adj[i].add(j)
        mst_adj[j].add(i)

    grid = assemble_rectangular_multi_seed(
        rotations, sigs, roles, border_sides,
        best_pairs, mst_adj,
        lam_mst=80.0, gamma=30.0,
        lam_size_h=40.0, lam_size_w=40.0,
        tol_h=2, tol_w=2
    )

    base = os.path.splitext(os.path.basename(path))[0]
    out_path = os.path.join(out_dir, f"{base}_mst_rect.png")
    canvas = render_grid_uniform(
        grid, rotations,
        out_path=out_path, show=show,
        title=f"MST rect: {base}"
    )
    print("Saved ->", out_path)
    return canvas"""

    base = os.path.splitext(os.path.basename(path))[0]
    out_path = os.path.join(out_dir, f"{base}_mst_rect.png")
    canvas = render_solution(
        solution, uprights,
        out_path=out_path, show=show,
        title=f"Solved: {base}"
    )
    print("Saved ->", out_path)
    
    if animate:
        anim_path = os.path.join(out_dir, f"{base}_animation.mp4")
        animate_puzzle_assembly(
            solution, uprights, piece_metadata,
            duration=anim_duration,
            out_path=anim_path
        )
    
    return canvas
```

```python
import json

def get_image_bounds(pieces):
    total_width = 0
    total_height = 0
    total_volume = 0
    for p in pieces:
        print(p.shape)
        total_width += p.shape[1]
        total_height += p.shape[0]
        total_volume += p.shape[1]*p.shape[0]
    side_length = math.sqrt(total_volume)
    return round(total_height/side_length), round(total_width/side_length), round(side_length)
    

def solve(pieces: list):
    """
    Solve puzzle using greedy spiral edge matching.
    
    Returns:
        List of [piece_idx, rotation, [row, col]] for each placed piece.
        rotation is 0 for translation-only (future: 0,1,2,3 for rotated puzzles)
    """
    total_difference = float('inf')
    ordered_pieces = []

    heightp, widthp, side_len = get_image_bounds(pieces)
    
    for i in range(len(pieces)):
        print("starting iteration: " + str(i))
        remaining = list(range(len(pieces)))  # Track indices, not images
        remaining.pop(i)
        pos = [0, 0]
        grid = np.full((int(side_len), int(side_len), 3), -1)
        rotation = 0  # Translation-only: no rotation
        temporder = [[i, rotation, pos]]  # [piece_idx, rotation, [row, col]]
        #direction = "r"
        #oppdir = "l"
        update_grid(pos, grid, pieces[i])
        pos = find_open_corner(grid)
        result = recursive_solver(i, remaining, pieces, temporder, pos, grid)

        if result[0] < total_difference:
            total_difference = result[0]
            ordered_pieces = result[1]

    return ordered_pieces


def recursive_solver(curr_idx, remaining, pieces, temporder, pos, grid):
    #currp = pieces[curr_idx]
    
    # Base case
    if len(remaining) <= 1:
        last_idx = remaining[0]
        p = pieces[last_idx]

        final_diff = 0
        if pos[0]>0:
            top_edge = p[0, :, :]
            final_diff += get_squared_difference(top_edge, grid[pos[0]-1, pos[1]:pos[1]+p.shape[1], :])
        if pos[1] > 0:
            left_edge = p[:, 0, :]
            final_diff += get_squared_difference(left_edge, grid[pos[0]:pos[0]+p.shape[0], pos[1]-1, :])
        if pos[1]+p.shape[1] < grid.shape[1]-1:
            right_edge = p[:, -1, :]
            final_diff += get_squared_difference(right_edge, grid[pos[0]:pos[0]+p.shape[0], pos[1]+p.shape[1], :])
        rotation = 0  # Translation-only
        neworder = temporder + [[last_idx, rotation, pos]]
        return [final_diff, neworder]

    #edge = return_edge(currp, direction)
    #matches = find_matching_edge(edge, pieces, oppdir, indices=remaining)

    matches = find_best_fill(grid, pos, remaining, pieces)
    
    best_diff = float('inf')
    sol = None

    if not matches:
        return [best_diff, sol]
    
    for i in range(min(1, len(matches))):
        match_idx = matches[i][2]  # piece index
        rotation = 0  # Translation-only
        neworder = temporder + [[match_idx, rotation, pos]]
        
        c = remaining.copy()
        c.remove(match_idx)
        
        g = grid.copy()
        #g[pos[0]][pos[1]] = 1
        update_grid(pos, g, pieces[match_idx])

        """
        # Direction changes (spiral logic)
        new_dir, new_oppdir = direction, oppdir
        if direction == 'r' and (pos[1]+1 >= g.shape[1] or g[pos[0]][pos[1]+1] == 1):
            new_dir, new_oppdir = 'b', 't'
        elif direction == 'b' and (pos[0]+1 >= g.shape[0] or g[pos[0]+1][pos[1]] == 1):
            new_dir, new_oppdir = 'l', 'r'
        elif direction == 'l' and (pos[1]-1 < 0 or g[pos[0]][pos[1]-1] == 1):
            new_dir, new_oppdir = 't', 'b'
        elif direction == 't' and g[pos[0]-1][pos[1]] == 1:
            new_dir, new_oppdir = 'r', 'l'
        """
        
        new_pos = find_open_corner(g)
        
        result = recursive_solver(match_idx, c, pieces, neworder, new_pos, g)
        if result[0] < best_diff:
            best_diff = result[0] + matches[i][0]
            sol = result[1]
    
    return [best_diff, sol]

def return_edge(piece, edge):
    if edge == "t":
        return piece[0, :, :]
    elif edge == "b":
        return piece[-1, :, :]
    elif edge == "r":
        return piece[:, -1, :]
    elif edge == "l":
        return piece[:, 0, :]
    return None

def update_pos(pos, direction, currp):
    if direction == "t":
        return [pos[0]-currp.shape[0], pos[1]]
    elif direction == "b":
        return [pos[0]+currp.shape[0], pos[1]]
    elif direction == "r":
        return [pos[0], pos[1]+currp.shape[1]]
    elif direction == "l":
        return [pos[0], pos[1]-currp.shape[1]]
    return None

def find_open_corner(grid):
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            if grid[i,j, 0] == -1: 
                return [i, j]

def update_grid(pos, grid, currp):
    grid[pos[0]:(pos[0]+currp.shape[0]), pos[1]:(pos[1]+currp.shape[1]), :] = currp
    
def update_pos_squares(pos, direction):
    if direction == "t":
        return [pos[0]-1, pos[1]]
    elif direction == "b":
        return [pos[0]+1, pos[1]]
    elif direction == "r":
        return [pos[0], pos[1]+1]
    elif direction == "l":
        return [pos[0], pos[1]-1]
    return None
    
def find_best_fill(grid, pos, indices, pieces):
    viable_pieces = []
    for i in indices:
        diff = 0
        p = pieces[i]
        if not can_fit(grid, pos, p):
            continue
        if(pos[0] > 0):
            grid_top_edge = grid[pos[0]-1, pos[1]:pos[1]+p.shape[1], :]
            piece_top_edge = p[0, :, :]
            diff += get_squared_difference(grid_top_edge, piece_top_edge)
        if(pos[1] > 0):
            edge_len = 0
            gpos = pos[0] 
            
            while gpos < grid.shape[0] and grid[gpos, pos[1]-1, 0]  != -1:
                edge_len += 1
                gpos += 1

            edge_len = min(edge_len, p.shape[0])
            grid_left_edge = grid[pos[0]:pos[0]+edge_len, pos[1]-1, :]
            piece_left_edge = p[:, 0, :]
            diff += get_squared_difference(grid_left_edge, piece_left_edge)
        viable_pieces.append([diff, p, i])

    viable_pieces.sort()
    return viable_pieces
    
def can_fit(grid, pos, piece):

    if piece.shape[0] + pos[0] > grid.shape[0] or piece.shape[1] + pos[1] > grid.shape[1]:
        return False
        
    grid_region = grid[pos[0]:pos[0]+piece.shape[0], pos[1]:pos[1]+piece.shape[1]]

    if np.any(grid_region != -1):
        return False
    
    return True

def find_matching_edge(edge, pieces, direction, indices=None):
    """
    Find best matching edges, optionally from a subset of pieces.
    
    Args:
        edge: Edge to match against
        pieces: List of all piece images
        direction: Direction for edge extraction
        indices: Optional list of piece indices to search (default: all)
    
    Returns:
        List of [difference, piece_image, piece_idx] sorted by difference
    """
    if indices is None:
        indices = range(len(pieces))
    
    edge_list = []
    for idx in indices:
        matching_edge = return_edge(pieces[idx], direction)
        diff = get_squared_difference(edge, matching_edge)
        edge_list.append([diff, pieces[idx], idx])

    edge_list.sort()
    return edge_list

def get_squared_difference(edge1, edge2):
    min_len = min(len(edge1), len(edge2))
    diff = edge1[:min_len].astype(np.int32) - edge2[:min_len].astype(np.int32)
    return np.sum(diff ** 2)
```
